## Set 

- Set will always have unique elements, even if you give repetitive elements it will just delete them
- Unlike dict, it won't have key, value pairs. It will just have elements.

- Set is an unordered collection data type that is iterable, immutable and has no duplicate elements.
- Advantage of using set over a list- it has highly optimized method of checking whether specific element is contained in list

```python
x = {1,2,3}
print(x)
x = {1,2,3,4,5,6,7,4,3,2}  # All the repetitions are gone, just have the unique ones.
print(x)
```

```python
my_set = {1.0, "hello", (1,2,3), (1.0, 2.0)}
print(my_set)
```

## Loops in Python

#### for loop and while loop..

##### In while loop we will write a condition, in terms of a variable and we make sure that condition becomes false

## Break and Continue Statements

```python
#Print out 0,1,2,3,4
#Break will come out of all the loop

count = 0
while True:
    print(count)
    count +=1
    if count >=5:
        break
print("*************")

#Print only odd numbers
#Continue comes out of current loop
for x in range(10):
    if x % 2 == 0:
        continue
    print(x)
```

## Comprehension in python

```python
alist = [1,2,3,4]
#multiply each element by 2
for i in alist:
    print(i*2)
```

```python
[i*2 for i in alist]
```

```python
words = "A quick brown fox jump into a cold river flowing".split()
print(words)
```

```python
stuff = [[w.upper(), w.lower(), len(w)] for w in words]
```

```python
stuff
```

```python
#print even numbers between 1 to 20
[i for i in range (1,20) if i%2==0]
```

```python
#list comprehension creates pythagorean triples:
[(x,y,z) for x in range (1,30) for y in range(x,30) for z in range (y,30) if x**2 + y**2 == z**2]
```

```python
#find intersection of 2 lists
x = [1,2,3]
y = [2,3,4,5]
[i for i in x if i in y]
```

```python
#Elements in first but not in second
x = [1,2,3]
y = [2,3,4,5]
[i for i in x if i not in y]
```

```python
#Elements in second but not in first
[val for val in y if val not in x]
```
